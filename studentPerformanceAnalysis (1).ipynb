```python
HeightNewDf = pd.DataFrame(df['Height(CM)'])
print(HeightNewDf)
Q1 = np.percentile(HeightNewDf['Height(CM)'], 25, interpolation = 'midpoint')
 
Q3 = np.percentile(HeightNewDf['Height(CM)'], 75, interpolation = 'midpoint')

IQR = Q3 - Q1
 
print("Old Shape: ", HeightNewDf.shape)
 
# Upper bound
upper = np.where(HeightNewDf['Height(CM)'] >= (Q3+1.5*IQR))

# Lower bound
lower = np.where(HeightNewDf['Height(CM)'] <= (Q1-1.5*IQR))
 
''' Removing the Outliers '''
HeightNewDf.drop(upper[0], inplace = True)
HeightNewDf.drop(lower[0], inplace = True)

print("New Shape: ", HeightNewDf.shape)

from scipy.stats import norm
plt.figure(figsize=(8, 6))
sns.distplot(HeightNewDf, fit=norm,color = 'r', kde=False)
plt.xlabel('Height')
plt.title("Student Height")
plt.show()
```

```python
# weight/(height in meter)^2

def changeValues(val):
  categories = ""
  if(val < 18.5):
    categories = 'Under weight'
  elif(val < 24.9):
    categories = 'Normal Weight'
  elif(val < 29.9):
    categories = 'Over Weight'
  elif(val < 34.9):
    categories = 'Obesity class 1'
  elif(val < 39.9):
    categories = 'Obesity class 2'
  elif(val > 40):
    categories = 'Obesity class 3'

  return categories


BMI = df[['Height(CM)','Weight(KG)']]

BMI['BMI'] = BMI['Weight(KG)']/((BMI['Height(CM)']*0.01)*2)

BMI['BMI'] = BMI['BMI'].apply(changeValues)

BMI['BMI'].value_counts()
```

```python
BMINewDf = pd.DataFrame(BMI['BMI'].value_counts())
BMINewDf['index'] = BMINewDf.index
BMINewDf.index = [x for x in range(0,6)]
plt.figure(figsize=(8, 8))
myexplodes = [0.1,0,0,0,0,0]
plt.pie(BMINewDf['BMI'],explode = myexplodes,wedgeprops={'linewidth':'2','edgecolor':'black'},textprops={'fontsize': 14},startangle=90, shadow=True,autopct='%1.1f%%')
plt.legend(labels = BMINewDf['index'])
plt.title('BMI')
plt.show()
```

```python
modelDataset = df[['college mark','10th Mark','12th Mark','daily studing time']]
dummy_variable = pd.get_dummies(modelDataset['daily studing time'])
modelDataset.drop('daily studing time',axis = 1,inplace = True)
modelDataset = pd.concat([modelDataset,dummy_variable],axis = 1)
modelDataset
```

```python
x = modelDataset.iloc[:,1:9]
y = modelDataset.iloc[:,:1]
print(x)
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.30,random_state=42)
ml = LinearRegression()
ml.fit(x_train,y_train)
```

```python
y_pred = ml.predict(x_test)
```

```python
print("Model Accuracy : ",r2_score(y_test,y_pred)*100)
```
