> Most people will easily get confused when it comes to the difference between **data science** and **data analytics**. So let's see the difference between **data science** and **data analytics** in detail.

**Data Analytics vs Data Science**

> * **Data analytics** involves answering questions generated for better business decision making. It uses existing information to uncover actionable data. Data analytics focuses on specific areas with specific goals.
* On the other hand, **data science** focuses on discovering new questions that you might not have realized needed answering to drive innovation. 
* Unlike **data analytics** which involves checking a hypothesis, **data science** tries to build connections and shapes the questions to answer them for the future. 
* If data science is a home for all the methods and tools, data analytics is a small room in that house. Data analytics is more specific and concentrated than data science.
* **Data analytics** focuses more on viewing the historical data in context while **data science** focuses more on machine learning and predictive modeling. 
* **Data science** is a multi-disciplinary blend that involves algorithm development, data inference, and predictive modeling to solve analytically complex business problems. 
* On the other hand, **data analytics** involves a few different branches of broader statistics and analysis.

# **Student's Height**

**The overall student's Height**

```python
# We know that a given dataset has some outlier values, so we are going to use the interquartile range (IQR) method 
# to remove those outlier values.

HeightNewDf = pd.DataFrame(df['Height(CM)'])

Q1 = np.percentile(HeightNewDf['Height(CM)'], 25, interpolation = 'midpoint')
 
Q3 = np.percentile(HeightNewDf['Height(CM)'], 75, interpolation = 'midpoint')

IQR = Q3 - Q1
 
# Upper bound
upper = np.where(HeightNewDf['Height(CM)'] >= (Q3+1.5*IQR))

# Lower bound
lower = np.where(HeightNewDf['Height(CM)'] <= (Q1-1.5*IQR))

print("Old Shape: ", HeightNewDf.shape)

''' Removing the Outliers '''
HeightNewDf.drop(upper[0], inplace = True)
HeightNewDf.drop(lower[0], inplace = True)

print("New Shape: ", HeightNewDf.shape)

from scipy.stats import norm
plt.figure(figsize=(8, 6))
sns.distplot(HeightNewDf['Height(CM)'], fit=norm,color = 'r', kde=False)
plt.xlabel('Height',fontsize = 20)
plt.title("Student Height",fontsize = 20)
plt.show()
```

From the above information, we could conclude that 90% of the student heights fall between 140 cm and 180 cm.

**Male students vs Female students, student's Height**

```python
maleHeightDf = male['Height(CM)']

femaleHeightDf = female['Height(CM)']

fig = plt.figure(figsize=(8, 6))
sns.distplot(maleHeightDf, fit=norm,color = 'r', kde=False,label = "Male")
sns.distplot(femaleHeightDf, fit=norm,color = 'b', kde=False,label = "Female")
plt.xlabel('Height',fontsize = 15)
plt.title("Student Height",fontsize = 15)
plt.legend()
plt.show()
```

# Student's Obesity(BMI) Level 

```python
BMIdf = pd.concat([male,female])
```

```python
# Let's assign each student a BMI value.

def changeValues(val):
  categories = ""
  if(val < 18.5):
    categories = 'Under weight'
  elif(val < 24.9):
    categories = 'Normal Weight'
  elif(val < 29.9):
    categories = 'Over Weight'
  elif(val < 34.9):
    categories = 'Obesity class 1'
  elif(val < 39.9):
    categories = 'Obesity class 2'
  elif(val > 40):
    categories = 'Obesity class 3'

  return categories


BMI = df[['Height(CM)','Weight(KG)']]

BMI['BMI'] = BMI['Weight(KG)']/((BMI['Height(CM)']*0.01)*2)

BMI['BMI'] = BMI['BMI'].apply(changeValues)

BMI['BMI'].value_counts()
```

```python
BMINewDf = pd.DataFrame(BMI['BMI'].value_counts())
BMINewDf['index'] = BMINewDf.index
BMINewDf.index = [x for x in range(0,6)]
plt.figure(figsize=(8, 8))
myexplodes = [0.05,0,0,0,0,0]
plt.pie(BMINewDf['BMI'],explode = myexplodes,wedgeprops={'linewidth':'2','edgecolor':'black'},textprops={'fontsize': 14},startangle=90, shadow=True,autopct='%1.1f%%')
plt.legend(labels = BMINewDf['index'],fontsize = 15)
plt.title('Students BMI Level',fontsize = 15)
plt.show()
```
